```python
from torchvision.utils import save_image
import torch
import time
from matplotlib import pyplot as plt 

from tools.file import mkdir
from utils.torch_loader_all import LoaderTest
from tools.toml import load_option
from app import  init, mask_op, array2image
```

```python
opt = load_option('../origin/test2.toml')
print(opt)
loader = LoaderTest(**opt)
```

```python
model_name = 'CSA2-1-1'
beta = 1
base_opt = load_option('../options/base2.toml')
model_opt = load_option('../options/train-new2.toml')
model = init(model_name, beta, model_opt, base_opt)

save_dir = f'D:/BaiduNetdiskWorkspace/result/measure/{model_name}'
mkdir(save_dir)
```

```python
valset = loader.valset()
maskset = loader.maskset()
```

```python
for image, mask in zip(valset, maskset):
    break
```

```python
def test(model, epoch, image, mask):
    model.load(epoch)
    image = image.cuda()
    mask = mask_op(mask)
    model.set_input(image, mask)
    model.set_gt_latent()
    model.test()
    real_A, real_B, fake_B = model.get_current_visuals()
    pic = (torch.cat([real_A, real_B, fake_B], dim=0) + 1) / 2.0
        
    save_image(pic, f"{save_dir}/epoch{epoch}.jpg", nrow=1)

    if epoch % 50 ==0:
        plt.figure(figsize=(6, 8))
        plt.subplot(1, 3, 1)
        x = array2image(real_A[0])
        plt.imshow(x)
        x = array2image(real_B[0])
        plt.subplot(1, 3, 2)
        plt.imshow(x)
        x = array2image(fake_B[0])
        plt.subplot(1, 3, 3)
        plt.imshow(x)
        plt.show()
```

```python
for k in range(100):
    test(model, k, image, mask)
```

```python
for k in range(100, 500):
    test(model, k, image, mask)
```
